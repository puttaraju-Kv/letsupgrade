```python
# Armstrong numbers in a range of (1042000 to 702648265) 
# exit the loop when print the fist armstrong number. 
```

```python

Anand = 1042000
Kumar = 702648265

for num in range(Anand, Kumar + 1):
   order = len(str(num)) 
   sum = 0

   temp = num
   while temp > 0:
       digit = temp % 10
       sum += digit ** order
       temp //= 10
       
   if num == sum:
       
       print(num)
       break
```
